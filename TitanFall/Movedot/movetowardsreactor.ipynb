```python
import os
from PIL import Image

def create_gif_from_frames_pillow(image_folder, output_gif_path, duration=100, loop=0):
    """
    Converts a sequence of image frames into an animated GIF using Pillow.

    Args:
        image_folder (str): The path to the directory containing the image frames.
        output_gif_path (str): The desired path and filename for the output GIF.
        duration (int, optional): The duration (in milliseconds) each frame is displayed.
                                  Defaults to 100ms (10 frames per second).
        loop (int, optional): The number of times the GIF should loop.
                              0 means it will loop indefinitely. Defaults to 0.
    """
    image_files = []
    for filename in sorted(os.listdir(image_folder)):
        if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):
            image_files.append(os.path.join(image_folder, filename))

    if not image_files:
        print(f"No image files found in '{image_folder}'. Please ensure your frames are there.")
        return

    images = []
    for file in image_files:
        try:
            img = Image.open(file)
            # Convert to RGB, as some image modes (e.g., RGBA) can cause issues with GIF
            if img.mode != 'RGB':
                img = img.convert('RGB')
            images.append(img)
        except Exception as e:
            print(f"Could not open image {file}: {e}")
            continue

    if not images:
        print("No valid images were loaded to create the GIF.")
        return

    # The first image saves the GIF, and subsequent images are appended.
    images[0].save(
        output_gif_path,
        save_all=True,
        append_images=images[1:],  # Append all images starting from the second one
        duration=duration,
        loop=loop,
        optimize=False # Set to True for smaller file size, but might lose some quality
    )
    print(f"GIF successfully created at: {output_gif_path}")

# Example Usage:
if __name__ == "__main__":
    # Create a dummy 'frames' folder and some images for demonstration
    if not os.path.exists('frames'):
        os.makedirs('frames')
    for i in range(5):
        img = Image.new('RGB', (100, 100), color=(i * 50, 255 - i * 50, 128))
        img.save(f'frames/frame_{i:03d}.png')

    create_gif_from_frames_pillow('frames', 'output_animation_pillow.gif', duration=300, loop=0)
```
